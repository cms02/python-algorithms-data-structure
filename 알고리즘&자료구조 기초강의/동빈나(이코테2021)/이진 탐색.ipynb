```python
# ì´ì§„ íƒìƒ‰ ì†ŒìŠ¤ì½”ë“œ êµ¬í˜„ (ì¬ê·€ í•¨ìˆ˜)
def binary_search(array, target, start, end):
    if start > end:
        return None
    mid = (start + end) // 2
    # ì°¾ì€ ê²½ìš° ì¤‘ê°„ì  ì¸ë±ìŠ¤ ë°˜í™˜
    if array[mid] == target:
        return mid
    # ì¤‘ê°„ì ì˜ ê°’ë³´ë‹¤ ì°¾ê³ ì í•˜ëŠ” ê°’ì´ ì‘ì€ ê²½ìš° ì™¼ìª½ í™•ì¸
    elif array[mid] > target:
        return binary_search(array, target, start, mid - 1)
    # ì¤‘ê°„ì ì˜ ê°’ë³´ë‹¤ ì°¾ê³ ì í•˜ëŠ” ê°’ì´ í° ê²½ìš° ì˜¤ë¥¸ìª½ í™•ì¸
    else:
        return binary_search(array, target, mid + 1, end)

# n(ì›ì†Œì˜ ê°œìˆ˜)ê³¼ target(ì°¾ê³ ì í•˜ëŠ” ê°’)ì„ ì…ë ¥ ë°›ê¸°
n, target = list(map(int, input().split()))
# ì „ì²´ ì›ì†Œ ì…ë ¥ ë°›ê¸°
array = list(map(int, input().split()))

# ì´ì§„ íƒìƒ‰ ìˆ˜í–‰ ê²°ê³¼ ì¶œë ¥
result = binary_search(array, target, 0, n - 1)
if result == None:
    print("ì›ì†Œê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤")
else:
    print(result + 1)
```

```python
# ì´ì§„ íƒìƒ‰ ì†ŒìŠ¤ì½”ë“œ êµ¬í˜„ (ë°˜ë³µë¬¸)
def binary_search(array, target, start, end):
    while start <= end:
        mid = (start + end) // 2
        # ì°¾ì€ ê²½ìš° ì¤‘ê°„ì  ì¸ë±ìŠ¤ ë°˜í™˜
        if array[mid] == target:
            return mid
        # ì¤‘ê°„ì ì˜ ê°’ë³´ë‹¤ ì°¾ê³ ì í•˜ëŠ” ê°’ì´ ì‘ì€ ê²½ìš° ì™¼ìª½ í™•ì¸
        elif array[mid] > target:
            end = mid - 1
        # ì¤‘ê°„ì ì˜ ê°’ë³´ë‹¤ ì°¾ê³ ì í•˜ëŠ” ê°’ì´ í° ê²½ìš° ì˜¤ë¥¸ìª½ í™•ì¸
        else:
            start = mid + 1
    return None

# n(ì›ì†Œì˜ ê°œìˆ˜)ê³¼ target(ì°¾ê³ ì í•˜ëŠ” ê°’)ì„ ì…ë ¥ ë°›ê¸°
n, target = list(map(int, input().split()))
# ì „ì²´ ì›ì†Œ ì…ë ¥ ë°›ê¸°
array = list(map(int, input().split()))

# ì´ì§„ íƒìƒ‰ ìˆ˜í–‰ ê²°ê³¼ ì¶œë ¥
result = binary_search(array, target, 0, n - 1)
if result == None:
    print("ì›ì†Œê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
else:
    print(result + 1)
```

# ê°’ì´ íŠ¹ì • ë²”ìœ„ì— ì†í•˜ëŠ” ë°ì´í„° ê°œìˆ˜ êµ¬í•˜ê¸°

```python
from bisect import bisect_left, bisect_right

# ê°’ì´ [left_value, right_value]ì¸ ë°ì´í„°ì˜ ê°œìˆ˜ë¥¼ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜
def count_by_range(a, left_value, right_value):
    right_index = bisect_right(a, right_value)
    left_index = bisect_left(a, left_value)
    return right_index - left_index

# ë°°ì—´ ì„ ì–¸
a = [1, 2, 3, 3, 3, 3, 4, 4, 8, 9]

# ê°’ì´ 4ì¸ ë°ì´í„° ê°œìˆ˜ ì¶œë ¥
print(count_by_range(a, 4, 4))
print(count_by_range(a, -1, 3))
```
